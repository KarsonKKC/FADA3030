```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pandas as pd
import numpy as np
```

```python
human_data = pd.read_csv("material list.csv")  # List of human-entered data
standardized_data = pd.read_csv("EPiC Database 2019.csv", skiprows=5)  # List of corresponding standardized data
```

```python
hd_df = human_data["Prompt"]
sd_df = standardized_data["Material"]
```

```python
# Tokenize your text data
tokenizer = Tokenizer(oov_token="<OOV>")  # Add OOV token
tokenizer.fit_on_texts(hd_df)
word_index = tokenizer.word_index
human_sequences = tokenizer.texts_to_sequences(hd_df)
human_sequences_padded = pad_sequences(human_sequences)

print("\nWord Index = " , word_index)
print("\nSequences = " , human_sequences)
print("\nPadded Sequences:")
print(human_sequences_padded)
```

```python
test_seq = tokenizer.texts_to_sequences(sd_df)
print("\nTest Sequence = ", test_seq)

padded = pad_sequences(test_seq, maxlen=10)
print("\nPadded Test Sequence: ")
print(padded)
```
